# Combining datasets
Pandas provides various facilities for easily combining together Series or DataFrames.
- **Concat:** combining DataFrames accross rows or columns
- **Join:** combining data on a key column or an index. Easier than merge if you only want to join on the index.
- **Merge:** combining data on common columns or indexes.

```python
from helpers import sample_df, hdisplay, nowrap_display
import pandas as pd
```

## Concat
Pandas .concat() method concatenates dataframe row-, or columnwise, <br>
with optional set logic (union or intersection) of the indexes on the other axis.

Concat simply stacks multiple dataframes together either vertically, or horizontally after aligning on row/column.

```python
# Sample data
left = sample_df("A0", "D3", prefix="L_")
right = sample_df("A0", "D3", prefix="R_")

hdisplay([left,right],["Left","Right"])
# built in function to display 
# display(left)
# display(right)
```

```python
hdisplay([
    pd.concat([left,right], axis="index"),
    pd.concat([left,right], axis="columns")],
    ["axis='index'","axis='columns'"]
    )
```

```python

hdisplay([
    # Resets the index so that it counts from 0 up. 
    pd.concat([left,right]).reset_index(drop=True),
    # Sets a column to be the index
    pd.concat([left,right]).set_index("A"),
    # Make new index built into the concat method
    pd.concat([left,right], axis="columns", ignore_index=True)],
    ["ignore_index=False","ignore_index=True","axis=columns","ignode_index=True"]
)
```

```python
# Sample data
left = sample_df("A0", "D3", prefix="L_")
right = sample_df("C2", "F5", prefix="R_")

hdisplay([left,right],["Left","Right"])
```

```python
hdisplay([
    pd.concat([left,right], axis="index"),
    pd.concat([left,right], axis="columns", join='inner'),  # intersection
    pd.concat([left,right], axis="columns", join='outer')], # union (default)
    ["'axis=index'","'axis='columns', join='inner'","'axis='columns'"], 20
)
```

```python
# Creates multi level index
hdisplay([
    pd.concat([left,right], axis='index', keys=['left','right']),
    pd.concat([left,right], axis='columns', keys=['left','right'])],
    ["axis='index'","axis='columns'"]
)

```

## Join
Join first aligns the index of two dataframes, and then picks up the remaining columns from the aligned rows of each data frame.

The **how** parameter can be any of the following:
- **left:** Get *all* rows from the left table, and join matching rows from the right table. (default)
- **right:** Get *all* rows from the right table, and join matching rows from the left table.
- **outer:** Get *all* rows from both tables, joining all matching rows on both sides. (union)
- **outer:** Get *only* rows that exist in both tables. (intersection)
- **cross:** Get *every* possible combination of rows from both tables. Length of new table = len(left) * len(right).


```python
# New sample data
left = sample_df("A0", "D3", prefix="L").add_prefix("L")
right = sample_df("C2", "F5", prefix="R").add_prefix("R")

hdisplay([left,right],["Left","Right"])
```

```python
# join how=left takes all the rows from the left table.
hdisplay([
    left.join(right, how="left"),   # default
    left.join(right, how="right"),
    left.join(right, how="inner"),  # intersection
    left.join(right, how="outer"), # union
    left.join(right, how="cross")], 
    ["how='left'","how='right'","how='inner'","how='outer'", "how='cross'"]
)
```

## Merge
Pandas .merge() method is similar to join, but is more versatile, and allows us to specify columns beside the index to join on for both dataframes. Works in exactly the same way as join in SQL.

Note: The *how* parameter works the same way as for join(), but the default for merge is "inner", not "left". 

```python
# New sample data
left = sample_df("A0", "D3").add_prefix("L")
left.loc[:,"LF"] = ["F10", "F11", "F12", "F13"]

right = sample_df("F10", "J13").add_prefix("R")

hdisplay([left,right],["Left","Right"])
```

```python
# None of these would have matched using join.
# Merge automatically finds the common columns, and aligns accordingly
# If no columns match, you can specify the columns to merge on.
left.merge(right, how="inner", left_on="LF", right_on="RF")
```

```python
employees = pd.read_json("../Data/employees.json")
departments = pd.read_json("../Data/departments.json")

display(employees)
display(departments)
```

```python
new_df = employees.merge(departments, left_on="department", right_on="department_name", sort=True).drop(columns=["department_name"])
new_df
```

```python
new_df.groupby("department").agg({"salary":"mean"}).sort_values("salary", ascending=False)
```
