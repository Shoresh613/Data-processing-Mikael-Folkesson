titanic.groupby("who").aggregate(
    number_of_passengers=pd.NamedAgg(column="survived", aggfunc="count"),
    number_of_survivors=pd.NamedAgg(column="survived", aggfunc="sum"),

```python
grouped = titanic.groupby('who').agg(
    number_of_passengers=pd.NamedAgg(column='who', aggfunc='count'),
    number_of_survivors=pd.NamedAgg(column='survived', aggfunc='sum'),
)

# Calculate the 'ratio_survivors' column
grouped['ratio_survivors'] = grouped['number_of_survivors'] / grouped['number_of_passengers']

# Reindex the DataFrame to match the desired index (0, 1, 2)
grouped = grouped.reindex(['man', 'woman', 'child'])

# # Reset the index to have a DataFrame with index 0, 1, 2
grouped = grouped.reset_index()

# # Rename the columns as needed
grouped = grouped.rename(columns={'who': 'Category'})

grouped
```

```python
def percent(column):
    return f"{(column/len(titanic)):.1%}"

grouped = titanic.groupby('who').agg(
    number_of_passengers=pd.NamedAgg(column='who', aggfunc='count'),
    number_of_survivors=pd.NamedAgg(column='survived', aggfunc='sum'),
)

# Calculate the 'ratio_survivors' column
grouped['ratio_survivors'] = grouped['number_of_survivors'] / grouped['number_of_passengers']

# Calculate the 'percentage_survived' column using the 'percent' function,
# as that function is not a method in the dataframe, but extra.
grouped['percentage_survived'] = grouped['number_of_survivors'].apply(percent)

# Reindex the DataFrame to match the desired index (0, 1, 2)
grouped = grouped.reindex(['man', 'woman', 'child'])

# Reset the index to have a DataFrame with index 0, 1, 2
grouped = grouped.reset_index()

# Rename the columns as needed
grouped = grouped.rename(columns={'who': 'Category'})

grouped
```
