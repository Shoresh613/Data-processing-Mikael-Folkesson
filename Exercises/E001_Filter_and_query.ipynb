## Exercises: Filter and query data

```python
import pandas as pd
autos = pd.read_json("../Data/autos.json")
autos
```

**Use the "autos" dataset to answer the following questions:**

**Q1:** How many cars from Jaguar are in the dataset, and what is the price of the most expensive one?

<details>
<summary>Answer</summary>
<br>
&nbsp;&nbsp;&nbsp;count=3
&nbsp;&nbsp;&nbsp;price=36000.0
</details>

```python
jaguars = autos.query("make == 'jaguar'")

if not jaguars.empty:
    print(f"Number of Jaguar in dataset: {len(jaguars)}")
    most_expensive_jaguar = jaguars.sort_values(by="price", ascending=False).iloc[0]
    print(f"Most expensive Jaguar: {most_expensive_jaguar['price']}")
else:
    print("No Jaguar cars found in the dataset.")
```

**T2:** Ask the user to input a brand, then print the price range for that brand.

<details>
<summary>Solution</summary>
<br>
&nbsp;&nbsp;&nbsp;<b>Example 1:</b><br>
&nbsp;&nbsp;&nbsp;Input the name of a brand: volvo<br>
&nbsp;&nbsp;&nbsp;The prices for cars of brand 'volvo' ranges from 12940.0 to 22625.0<br>
<br>
&nbsp;&nbsp;&nbsp;<b>Example 2:</b><br>
&nbsp;&nbsp;&nbsp;Input the name of a brand: toyota<br>
&nbsp;&nbsp;&nbsp;The prices for cars of brand 'toyota' ranges from 5348.0 to 17669.0<br>
<br>
&nbsp;&nbsp;&nbsp;<b>Example 3:</b><br>
&nbsp;&nbsp;&nbsp;Input the name of a brand: tesla<br>
&nbsp;&nbsp;&nbsp;The brand 'tesla' does not exists in the dataset.<br>
</details>

```python
print("Get price range of car brand. Enter emptry string to exit.")
print("==========================================================\n")

while True:
    try:
        brand = input("Enter brand: ").lower()
        if brand == "":
            break
        if len(autos.query("make == @brand")) == 0:
            raise ValueError
        else:
            break
    except ValueError:
        print("Brand not in list of cars. Try again, or enter empty string to exit.")

if brand != "":
    brandcars = autos.query("make == @brand")
    most_expensive_car = brandcars.sort_values(by="price", ascending=False).iloc[0]
    least_expensive_car = brandcars.sort_values(by="price", ascending=True).iloc[0]

    print(f"The price range of {brand.title()} cars is: {least_expensive_car['price']} - {most_expensive_car['price']} USD.")
```

**T4:** Ask the user to input a brand, then export all cars of that brand to a .csv file with the same name as the brand.

<details>
<summary>Solution</summary>
<br>
&nbsp;&nbsp;&nbsp;<b>Example 1:</b><br>
&nbsp;&nbsp;&nbsp;Input the name of a brand: volkswagen<br>
&nbsp;&nbsp;&nbsp;Exported 12 cars to 'volkswagen.csv'<br>
<br>
&nbsp;&nbsp;&nbsp;<b>Example 2:</b><br>
&nbsp;&nbsp;&nbsp;Input the name of a brand: tesla<br>
&nbsp;&nbsp;&nbsp;The brand 'tesla' does not exists in the dataset.<br>
</details>

```python
print("Export all cars of brand.\nEnter emptry string to exit.")
print("============================================================================\n")

while True:
    try:
        brand = input("Enter brand: ").strip().lower()
        if brand == "":
            break
        if len(autos.query("make == @brand")) == 0:
            raise ValueError
        else:
            break
    except ValueError:
        print("Brand not in list of cars. Try again, or enter empty string to exit.")

if brand != "":
    brandcars = autos.query("make == @brand")
    try:
        brandcars.to_csv(f"'E001_export_{brand}.csv'")
        print(f"Saved file: E001_export_{brand}.csv")
    except:
        print("Could not save csv file.")
    
```
